```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 直方图是一种像素的统计数据 的图形表示
# 横坐标像素亮度
# 纵坐标 对应亮度的像素的个数
# 归一化的直方图 纵坐标代表出现概率
'''
• images：输入的图像的指针；
• nimages：输入图像个数；
• channels：需要统计直方图的第几通道；
• mask：掩模，mask必须是一个8位（CV_8U）的数组并且和images的数组大小相同；
• hist：直方图计算的输出值；
• dims：输出直方图的维度（由channels指定）；
• histSize：直方图中每个dims维度需要分成多少个区间（如果把直方图看作一个一个竖条的话，就是竖条的个数）；
• ranges：统计像素值的区间；

'''
pass
```

```python
# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) -> hist

img = cv2.imread('../data/linna.jpg')

hist = cv2.calcHist(images = [img],channels = [0,1,2],mask = None,\
                    histSize = [256,256,256],ranges=[0,255,0,255,0,255])

# hist_b,hist_g,hist_r = hist[:,:,::1]
```

```python
# 不用opencv的统计方法(比opencv慢)

img = cv2.imread('../data/linna.jpg',0) 
plt.hist(img.ravel(),bins=256,range=[0,255])# 只读灰度图
```

```python
# 绘制直方图
img = cv2.imread('../data/linna.jpg')

hist_b = cv2.calcHist(images = [img],channels = [0],mask = None,\
                    histSize = [256],ranges=[0,255])

hist_g = cv2.calcHist(images = [img],channels = [1],mask = None,\
                    histSize = [256],ranges=[0,255])

hist_r = cv2.calcHist(images = [img],channels = [2],mask = None,\
                    histSize = [256],ranges=[0,255])
plt.plot(hist_b,color='b',label = 'blue')
plt.plot(hist_g,color='g',label = 'green')
plt.plot(hist_r,color='r',label = 'red')
plt.legend()
```

```python
# 统计直方图可以使用mask
# mask是一张黑白图,白色的位置代表想要进行统计的区域
mask = np.zeros(img.shape[:2],dtype = np.uint8)
mask[::2,:] = [255] # 隔行设置一行白色
cv2.imshow('mask',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()

img = cv2.imread('../data/linna.jpg')

hist_b = cv2.calcHist(images = [img],channels = [0],mask = mask,\
                    histSize = [256],ranges=[0,255])

hist_g = cv2.calcHist(images = [img],channels = [1],mask = mask,\
                    histSize = [256],ranges=[0,255])

hist_r = cv2.calcHist(images = [img],channels = [2],mask = mask,\
                    histSize = [256],ranges=[0,255])
plt.plot(hist_b,color='b',label = 'blue')
plt.plot(hist_g,color='g',label = 'green')
plt.plot(hist_r,color='r',label = 'red')
plt.legend()
# 隔行采样对直方图分布没什么影响
```
