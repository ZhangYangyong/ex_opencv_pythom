```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('../data/001.jpg')
img2 = cv2.imread('../data/002.jpg')
```

# 四则运算

```python
# 展示一下原始图片
show_img = np.hstack([img1,img2])
show_img =cv2.resize(show_img,(0,0),fx=0.1,fy =0.1)
plt.imshow(cv2.cvtColor(show_img,cv2.COLOR_BGR2RGB))
```

```python
# 加法运算 cv2.add(src1, src2[, dst[, mask[, dtype]]]) -> dst
# 相加操作要求两张图片维度相同 长宽,通道
print(img1.shape,img2.shape)

add_img = cv2.add(img1,img2) # add 操作按照numpy数值相加,大于255的值取255
add_img =cv2.resize(add_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./add_img.png',cv2.cvtColor(add_img,cv2.COLOR_BGR2RGB))
#  ![add_img](./add_img.png)
```

![add_img](./add_img.png)

```python
# 图片不光可以和图片相加,也可以和单个数字或者向量相加,广播 利用 + 号运算符 这种最后的结果是255取余
new_img = img1 +100
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./add_img2.png',cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![add_img2](./add_img2.png)

```python
# 只对单个通道处理
new_img = img1 +np.array((0,0,100),dtype=np.uint8)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./add_img3.png',cv2.cvtColor(add_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![add_img3](./add_img3.png)

```python
# 减法操作 subtract(src1, src2[, dst[, mask[, dtype]]]) -> dst
# 亦可以数字,向量 相减 - 运算符
subtract_img = cv2.subtract(img1,img2) # 小于0变成0
subtract_img =cv2.resize(subtract_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./subtract_img.png',cv2.cvtColor(add_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(subtract_img,cv2.COLOR_BGR2RGB))
```

![subtract_img](./subtract_img.png)

```python
new_img = img1 -100 # 小于-1 = 255 ,-2 = 254 ...
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./subtract_img2.png',cv2.cvtColor(add_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![subtract_img2](./subtract_img2.png)

```python
# 四则运算 实际用途比较小
# 按权重相加叠加图片用途更广 addWeighted(src1, alpha, src2, beta, gamma[, dst[, dtype]]) -> dst 
new_img = cv2.addWeighted(img1,0.8,img2,0.2,30)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./addWeighted.png',cv2.cvtColor(add_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

# 位运算 与或非 异或

```python
# 非运算 0->255 0000 0000 -> 1111 1111  125 =0111 1101 -> 1000 0010=130 等价于 255-125 和摄影中的反相一样
a = np.array([[120]],np.uint8)
a_not = cv2.bitwise_not(a)
print(a_not)
new_img = cv2.bitwise_not(img1)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./bitwise_not.png',cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![bitwise_not](./bitwise_not.png)

```python
# 与运算
# 128 and 1 1000 0000 and 0000 0000 1->0000 0000
a = np.array([[128]],np.uint8)
b = np.array([[1]],np.uint8)
a_and_b = cv2.bitwise_and(a,b)
print(a_and_b)
```

```python
# 与运算会保留更多的0 整体图片会变暗(不大于两个像素中较小的那个,最多相等)
new_img = cv2.bitwise_and(img1,img2)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./bitwise_and.png',cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![bitwise_and](./bitwise_and.png)

```python
# 或运算 
a = np.array([[128]],np.uint8)
b = np.array([[1]],np.uint8)
a_or_b = cv2.bitwise_or(a,b)
print(a_or_b)
```

```python
# 或运算会保留更多的1 整体图片会变亮(不小于两个像素中较大的那个,最多相等)
new_img = cv2.bitwise_or(img1,img2)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./bitwise_or.png',cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```

![bitwise_or](./bitwise_or.png)

```python
# 异或
# 1001 1001  153
# 1011 0101  181 
# 0010 1100  44
a = np.array([[153]],np.uint8)
b = np.array([[181]],np.uint8)
a_xor_b = cv2.bitwise_xor(a,b)
print(a_xor_b)
```

```python
#异或运算
'''
图像加密解密 - 可以使用异或运算对图像进行简单的加密,与密钥进行异或后得到密文图像,解密时再与密钥异或可以还原原图。
'''
new_img = cv2.bitwise_xor(img1,img2)
new_img =cv2.resize(new_img,(0,0),fx=0.1,fy =0.1)
plt.imsave('./bitwise_xor.png',cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
# plt.imshow(cv2.cvtColor(new_img,cv2.COLOR_BGR2RGB))
```
